```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('indian_food.csv')
df.head()
```

```python
df.shape
```

```python
df2=df.cook_time.mean()
```

We have some 'NAN' value in cook_time column so we replace this with mean of cook_time

```python
df.fillna(df2)
df.head()
```

in our case we don't need  course column so we delete it

```python
df3=df.drop(columns='course')
df3.head()
```

```python
df3.groupby('flavor_profile').flavor_profile.count()
```

```python
a,b,c,d,f=df3.groupby('flavor_profile').flavor_profile.count()
```

```python
labels = 'bitter', 'sour','spicy','sweet'
sizes = [ b,c,d,f]
colors = ['gold', 'lightskyblue' ,'lightcoral', 'lightskyblue']
explode = (0.5, 0,0,0) 

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
